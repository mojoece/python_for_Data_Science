Loops
- For, while and do while
- Loops allow us to iterate
- Loops helps us iterate until a condition is specified

```python
seq = [1, 2, 3, 4, 5]
```

```python
for item in seq:  # The temporary variable name item can be whatever we want it to be
    print(item)   # Here, in for loop, the iteration happens based on what we write in the loop intro and not the body 
```

```python
i = 1
while i < 5:
    print(f'i is: {i}')
    i = i + 1    # Condition is written inside the body unlike the for loop
```

```python
for i in range(1, 11):
    print(f'12 x {i} = {12*i}')
```

```python
list(range(0, 5))
```

```python
list(range(10))
```

List Compherension

```python
x = [1, 2, 3, 4]
```

```python
out  = []

for num in x:
    out.append(num**2)    # List comphernsion is a shortcut for these types of problems as this is widely used
```

```python
print(out)
```

```python
out = [num**3 for num in x]    # This is list compherension and the difference from previous approach is that it is short and written backwards
```

```python
print(out)
```

Functions
- Helps us to not continually write same sort of code again and again.
- Instead, we can write everything inside a function and in the end, call that function itself

```python
def my_fun(name = 'Default Name'):  # When we dont give any input then the default paramater gets executed
    print('Hello ' + name + '. I hope you are having a great day.')
```

```python
my_fun('mojo')
```

```python
my_fun   # Tells us about the object instead of calling it
```

```python
def square(num):
    """
    THIS IS A DOCSTRING.
    CAN GO MULTIPLE LINES.
    THIS FUNCTION SQUARES A NUMBER.
    """
    return num**2  # This will return the square rather than just printing it. This will get executed at the end of the function and this value will get returned from the functio
```

```python
output = square(5)
```
