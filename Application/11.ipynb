Dictionaries
- In key-value pair rather than indexing
- Similar to hash maps
- Mutable in nature

```python
d = {
    'key1' : 'value',
    'key2' : 123
}
```

```python
d['key1']
```

```python
d['key2']
```

```python
d = {
    'k1': [1, 2, 3] # With this type of case where list is shown as a value, we can use the normal indexing to access the elements
}
```

```python
d['k1']
```

```python
d['k1'][1]
```

```python
my_list = d['k1']
```

```python
my_list[1]    
```

```python
d = {
    'k1': {
        'name':'k1 buddy',
        'class': 2,
        'place': 'nothing',
        'marks': [1, 2, 3, 4]
    }
}
```

```python
d['k1']
```

```python
d['k1']['marks'][2]
```

```python
# Dictionaries do not retain any order, they are just carrying key value pairs
```

Tuples
- Similar to Lists but instead of square brackets, we have normal round parenthesis
- We can use indexing here
- List elements can be reassigned but tuples don't allow that as they are `Immutable` and cant be altered or do not support item assignment

```python
t = (1, 2, 3)
```

```python
t[0]
```

```python
t[0] = 'trying to change tuple' # Gives error as tuple is immutable
```

Sets
- Collection of unique elements (it 1 element multiple times then also unique will show)
- Looks similar to dictionary due to curly braces but it doesn't have colons rather just elements are there

```python
{1, 2, 3, 4}
```

```python
{1, 2, 3, 5, 2, 1, 2, 1, 1, 2, 4, 5, 7, 23}
```

```python
# Using a list along with set to grab unique elements

set([1, 2, 1, 1, 1, 3, 4, 5, 5, 2, 3, 4, 34])
```
