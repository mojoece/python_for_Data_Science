# Missing Data

Convenient methods to deal with Missing Data in pandas:

```python
import numpy as np
import pandas as pd
```

```python
d = {'A':[1,2,np.nan],
     'B':[5,np.nan,np.nan],
     'C':[1,2,3]}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df.dropna(axis = 0)      # DataFrame or None
# DataFrame with NA entries dropped from it or None if inplace=True.
```

```python
df.dropna(axis = 1)
```

```python
df.dropna(thresh = 2)      # removes rows with na values more than 2
```

```python
df.fillna('Filled Value')
```

```python
df
```

```python
df['A'].fillna(value=df['A'].mean())
```
