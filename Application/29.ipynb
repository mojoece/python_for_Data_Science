## Multi-Index and Index Hierarchy

Let us go over how to work with Multi-Index, first we'll create a quick example of what a Multi-Indexed DataFrame would look like:

```python
import numpy as np
import pandas as pd
from numpy.random import randn
np.random.seed(101)
```

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
outside
```

```python
inside
```

```python
hier_index
```

```python
df = pd.DataFrame(randn(6,2), hier_index, ['A','B'])
```

```python
df
```

```python
# Now let's show how to index this! For index hierarchy we use df.loc[], if this was on the columns axis, you would just use normal bracket notation df[]. Calling one level of the index returns the sub-dataframe:
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df.index.names
```

```python
df.index.names = ['Groups','Num']
```

```python
df
```

```python
df.loc['G2'].loc[2]['B']
```

```python
# Returns a cross-section(Rows or Columns) from the Series/DataFrame. Defaults to cross-section on the rows(axis = 0)
df.xs   # Used when we have a multi level index
```

```python
df.xs('G1')
```

```python
df.xs(1,level='Num') # Useful to directly get the data placed there
```
