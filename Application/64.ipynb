# World Choropleth Map

Now let's see an example with a World Map:

```python
df = pd.read_csv('src/63/2014_World_GDP')
```

```python
df.head()
```

```python
data = dict(
    type = 'choropleth',
    locations = df['CODE'],
    z = df['GDP (BILLIONS)'],
    text = df['COUNTRY'],
    colorbar = {'title':'GDP in Billions USD'}
)
```

```python
layout = dict(
    title = '2014 Global GDP',
    geo = dict(
        showframe = False, 
        projection = {'type':'natural earth'})
)
```

```python
choromap = go.Figure(data = [data], layout = layout)
```

```python
iplot(choromap)
```
