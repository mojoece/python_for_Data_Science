# DataFrames

DataFrames are the workhorse of pandas and are directly inspired by the R programming language. We can think of a DataFrame as a bunch of Series objects put together to share the same index. Let's use pandas to explore this topic!

```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn
```

```python
np.random.seed(101)    # We get the ame random number using this
```

```python
df = pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
```

```python
df      # Here, Dataframs is a bunch of series that hares a common index
```

```python
df['W']
```

```python
type(df['W'])
```

```python
type(df)
```

```python
# SQL Syntax (NOT RECOMMENDED!)
df.W
```

```python
# Pass a list of column names
df[['W','Z']]
```

```python
df['new'] = df['W'] + df['Y']
```

```python
df
```

```python
df.drop('new', axis = 1)      # By default, the axis is 0 i.e. it refers to
# the rows/indexes. For columns, we need to change the axis to 1
```

```python
# Not inplace unless specified!
df
```

```python
df.drop('new', axis = 1, inplace=True)    # This is required to specifically
# Tell that, we want to delete this thing
```

```python
df
```

```python
df.drop('E', axis = 0, inplace=True)
```

```python
df
```

```python
df.shape     # Reason for the axis attribute being either 0 or 1
```

```python
# ** Selecting Rows**
df.loc['C']   # Not only the column are series but all the rows are series as well
```

```python
# Or select based off of position instead of label 
df.iloc[2]          # This should be used when we want to grab the row series 
# Using indexes rather than the labels
```

```python
# ** Selecting subset of rows and columns **
df.loc['B', 'Y']
```

```python
df.loc[['A', 'B'], ['W', 'Y']]
```
