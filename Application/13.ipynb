map and filter
- map is a builtin function in python that helps us to create a iterator automatically and then output the values based on that

```python
def times2(num):
    return num*2
```

```python
times2(5)
```

```python
seq = [1, 2, 3, 4, 5]
```

```python
map(times2, seq)
```

```python
list(map(times2, seq))
```

```python
# In this map, we dont want to write the separate big function for using it so, what we do instead is that we use lambda functions for this
```

lambda function or anonymous function
- Here, we don't want to define an entire function rather things become much simpler and with the information that we need.

```python
test = lambda var: var*2
```

```python
test(3)
```

```python
list(map(lambda var: var*3, seq))
```
