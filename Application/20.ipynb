## Bracket Indexing and Selection
The simplest way to pick one or some elements of an array looks very similar to python lists:

```python
#Get a value at an index
arr[8]
```

```python
#Get values in a range
arr[1:5]
```

```python
arr[3:]
```

```python
arr[:4]
```

```python
arr[:6:2]
```

```python
arr[2:10:2]
```

```python
arr[::3]
```

## Broadcasting

Numpy arrays differ from a normal Python list because of their ability to broadcast:

```python
#Setting a value with index range (Broadcasting)
arr[0:5] = 100
```

```python
arr
```

```python
arr = np.arange(0, 11)
```

```python
arr
```

```python
slice_of_arr = arr[0:6]
```

```python
slice_of_arr
```

```python
slice_of_arr[:]
```

```python
slice_of_arr[:] = 99
```

```python
slice_of_arr
```

Now note the changes also occur in our original array!

```python
arr         # Here, we can observe that by changing the values in slice_of_arr, the values automatically gets changed in the main array.
# This means that here, the data is not copied separately rather it is just a view of the original array. 
# Numpy does this to avoid memory issues with very large arrays
```

Data is not copied, it's a view of the original array! This avoids memory problems!

```python
# For the copy, what we can do is separately make a copy of the array
arr_copy = arr.copy()
```

```python
arr_copy
```

```python
arr
```

```python
arr_copy[0:6] = np.arange(0,6)
```

```python
arr_copy
```

## Indexing a 2D array (matrices)

The general format is **arr_2d[row][col]** or **arr_2d[row,col]**. I recommend usually using the comma notation for clarity.

```python
arr_2d = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
```

```python
arr_2d
```

```python
#Indexing row
arr_2d[0]
```

```python
# Format is arr_2d[row][col] or arr_2d[row,col]

# Getting individual element value
arr_2d[0][0]
```

```python
# Getting individual element value
arr_2d[0,0]
```

```python
# 2D array slicing

#Shape (2,2) from top right corner
arr_2d[1:,1:]
```

```python
arr_2d[0:2,1:3]
```

```python
arr_2d[:2,1:]
```

```python
#Shape bottom row
arr_2d[2]
```

```python
#Shape bottom row
arr_2d[2,:]
```

```python
arr_2d[2,0:3]
```

### Fancy Indexing

Fancy indexing allows you to select entire rows or columns out of order,to show this, let's quickly build out a numpy array:

```python
#Set up matrix
arr2d = np.zeros((10,10))
```

```python
#Length of array
arr_length = arr2d.shape[1]
```

```python
#Set up array

for i in range(arr_length):
    arr2d[i] = i
    
arr2d
```

```python
arr2d[[2,4,6,8]]
```

```python
#Allows in any order
arr2d[[6,4,2,7]]
```

## Conditional Selection
To use brackets for selection based off of comparison operators.

```python
narr = np.arange(1, 11)
```

```python
narr
```

```python
bool_arr = narr > 5
```

```python
bool_arr
```

```python
narr[bool_arr]       # Got the values that has only true in this.
```

```python
# We can also do it directly in one single step i.e.
narr[narr > 5]
```

```python
narr[narr < 3]
```

```python
arr_2d = np.arange(50).reshape(5,10)
```

```python
arr_2d
```

```python
arr_2d[1:4,5:9]
```
